```python
## Generate Traits

TOTAL_IMAGES = 2000 # Number of random unique images we want to generate

all_images = [] 

# A recursive function to generate unique image combinations
def create_new_image():
    
    new_image = {} #

    # For each trait category, select a random trait based on the weightings 
    new_image ["Background"] = random.choices(Background, background_weights)[0]
    new_image ["Wing"] = random.choices(Wing, wing_weights)[0]
    new_image ["Design"] = random.choices(Design, design_weights)[0]
    new_image ["Base"] = random.choices(Base, base_weights)[0]
    
    if new_image in all_images:
        return create_new_image()
    else:
        return new_image
    
    
for i in range(TOTAL_IMAGES): 
    
    new_trait_image = create_new_image()
    
    all_images.append(new_trait_image)
    
```

```python
# Returns true if all images are unique
def all_images_unique(all_images):
    seen = list()
    return not any(i in seen or seen.append(i) for i in all_images)

print("Are all images unique?", all_images_unique(all_images))
```

```python
# Add token Id to each image
i = 0
for item in all_images:
    item["tokenId"] = i
    i = i + 1
    
```

```python
#### Generate Images    
for item in all_images:
    
    
    im1 = Image.open(f'./trait-layers/Block Butterflies/background/{background_files[item["Background"]]}.png').convert('RGBA')
    im2 = Image.open(f'./trait-layers/Block Butterflies/wing/{wing_files[item["Wing"]]}.png').convert('RGBA')
    im3 = Image.open(f'./trait-layers/Block Butterflies/design/{design_files[item["Design"]]}.png').convert('RGBA')
    im4 = Image.open(f'./trait-layers/Block Butterflies/base/{base_files[item["Base"]]}.png').convert('RGBA')


    #Create each composite
    com1 = Image.alpha_composite(im1, im2)
    com2 = Image.alpha_composite(com1, im3)
    com3 = Image.alpha_composite(com2, im4)

    
    #Convert to RGB
    rgb_im = com3.convert('RGB')
    file_name = str(item["tokenId"]) + ".png"
    rgb_im.save("./images/" + file_name)
    
    
    
```

```python
#### Generate Metadata for each Image    

f = open('./metadata/all-traits.json',) 
data = json.load(f)


IMAGES_BASE_URI = "ADD_IMAGES_BASE_URI_HERE"
PROJECT_NAME = "Block Butterflies"

def getAttribute(key, value):
    return {
        "trait_type": key,
        "value": value
    }
for i in data:
    token_id = i['tokenId']
    token = {
        "image": IMAGES_BASE_URI + str(token_id) + '.png',
        "tokenId": token_id,
        "name": PROJECT_NAME + ' ' + str(token_id),
        "attributes": []
    }
    token["attributes"].append(getAttribute("Background", i["Background"]))
    token["attributes"].append(getAttribute("Base", i["Base"]))
    token["attributes"].append(getAttribute("Wing", i["Wing"]))
    token["attributes"].append(getAttribute("Design", i["Design"]))
    token["attributes"].append(getAttribute("Design Color", i["Design Color"]))

    with open('./metadata/' + str(token_id), 'w') as outfile:
        json.dump(token, outfile, indent=4)
f.close()
```
